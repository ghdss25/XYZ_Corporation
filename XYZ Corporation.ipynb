## Respostas 

## 1 - Qual a Categoria obteve a maior taxe de despesas na empresa

```python
despesas_empresa['Valor'] = pd.to_numeric(despesas_empresa['Valor'], errors='coerce')
```

```python
categoria_despesas = despesas_empresa.groupby('Categoria')['Valor'].sum()
```

```python
categoria_maior_despesa = categoria_despesas.idxmax()
valor_maior_despesa = categoria_despesas.max()

print(f"A categoria com a maior taxa de despesas é '{categoria_maior_despesa}' com um valor total de R$ {valor_maior_despesa:.2f}")
```

## 2 - Informe o total e a média das despesas das Categorias

```python
categoria_despesa_total = despesas_empresa.groupby('Categoria')['Valor'].sum().rename("Total de Despesas") 
categoria_despesa_media = despesas_empresa.groupby('Categoria')['Valor'].mean().rename("Média de Despesas") 

analise_despesa = pd.DataFrame({"Total de Despesas" : categoria_despesa_total, 
                                "Média de Despesas" : categoria_despesa_media}) 

analise_despesa["Total de Despesas"] = analise_despesa["Total de Despesas"].map(lambda x: "R$ {:,.2f}".format(x))
analise_despesa["Média de Despesas"] = analise_despesa["Média de Despesas"].map(lambda x: "R$ {:,.2f}".format(x))

analise_despesa
```

## 3 - Defina qual dia e o mês que teve maior despesas nas Categorias

```python
despesas_empresa['Data'] = pd.to_datetime(despesas_empresa["Data"])
despesas_empresa['Ano'] = despesas_empresa['Data'].dt.year
despesas_empresa['Mês'] = despesas_empresa['Data'].dt.month
despesas_empresa['Dia'] = despesas_empresa['Data'].dt.day
```

```python
despesas_empresa
```

```python
categoria_max_despesa = despesas_empresa.groupby('Categoria').apply(lambda x: x.loc[x['Valor'].idxmax()])
categoria_max_despesa['Valor'] = categoria_max_despesa['Valor'].apply(lambda x: f'R$ {x:.2f}')
```

```python
# Ordenar os resultados pelo maior valor
categoria_max_despesa = categoria_max_despesa.sort_values(by='Valor', ascending=False)
```

```python
display(categoria_max_despesa[['Dia', 'Valor']])
```

```python
display(categoria_max_despesa[['Mês', 'Valor']])
```

##  4 - Defina qual dia e o mês que teve menor despesas nas Categorias

```python
categoria_min_despesa = despesas_empresa.groupby('Categoria').apply(lambda x: x.loc[x['Valor'].idxmin()])
categoria_min_despesa['Valor'] = categoria_min_despesa['Valor'].apply(lambda x: f'R$ {x:.2f}')
```

```python
# Ordenar os resultados pelo menor dia
categoria_min_despesa = categoria_min_despesa.sort_values(by='Dia', ascending=True)
```

```python
display(categoria_min_despesa[['Dia', 'Valor']])
```

```python
# Ordenar os resultados pelo menor mês
categoria_min_despesa = categoria_min_despesa.sort_values(by='Mês', ascending=False)
```

```python
display(categoria_min_despesa[['Mês', 'Valor']])
```
